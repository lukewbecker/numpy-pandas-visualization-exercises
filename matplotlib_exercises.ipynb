## Lesson matplotlib Exercises

```python
import matplotlib.pyplot as plt
import random
import math
```

```python
# Use matplotlib to plot the following equation:
# y = (x ** 2) - x + 2

x = list(range(-20, 20))
y = [i ** 2 - i + 2 for i in x]


plt.plot(x, y)
plt.annotate('Origin', xy=(0,0), xytext = (0, 100),
            arrowprops=dict(facecolor = 'darkorange'))
plt.title("$y = x^2 - x + 2$")
plt.grid(b=True)
plt.show()
```

#### Equation 2
- $y = x^3$

```python
x3 = list(range(-20, 24))
y3 = [y ** 3 for y in x3]

plt.plot(x3, y3)
plt.savefig('cubed_figure.pdf')
plt.ylabel('$x^3$')
plt.xlabel('$x$')
plt.title("$y = x^3$")
plt.grid(b=True)
plt.show()
```

#### Equation 3
- $y = 2^x$

```python

x4 = list(range(-7, 8))
y4 = [2 ** x for x in x4]

plt.plot(y4, c='firebrick')
plt.ylabel("$2^x$", c="rebeccapurple")
plt.xlabel("$x$", c="sienna")
plt.grid(b=True)
plt.show()
```
