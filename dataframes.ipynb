```python
# What is the the weight of the fastest animal?

Mammals.sort_values(by = 'speed', ascending = False).head(1)
```

```python
# What is the overal percentage of specials?

# Had to find the number of true within the column by using vectorized operations...
# .shape provides the number of rows and the number of columns. However, since I'm calling a specific column, .shape will only return the number of rows, which is what I need to do my pct equation. Thanks to my classmates who reminded me of that. Took me a while to figure that part out.
# Then, using value_counts allows me to break up the True/False booleans.
# Finally, I can use .tail(1) (in this instance) to only return the last entry (of 2 total entries) 
# showing the overall percentage of specials is 0.093458, or a little over 9%.

(Mammals.specials.value_counts() / Mammals.specials.shape).tail(1)
```

```python
# How many animals are hoppers that are above the median speed? What percentage is this?

# First, find the median speed using the median operator for all data values in the speed column:
median_speed = Mammals.speed.median()

# Now, I have to use a True/False boolean mask(s) to figure out of a given row is above or below the median speed, AND if it is a hopper.

# I opted to filter multiple columns within the dataframe, and use the "&" operator within the code 
# to return the new dataframe showing hoppers who's speed is above the median speed of 48.0.

fast_hoppers = Mammals[(Mammals['speed'] > median_speed) & (Mammals['hoppers'] == True)]
fast_hoppers
```
